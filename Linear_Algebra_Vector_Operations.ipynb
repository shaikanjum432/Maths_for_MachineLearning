```python
plt.quiver(0,0,4,5)
```

## Addition of two Vectors

```python
vector1 = np.asarray([0,0,2,3])
vector2 = np.asarray([0,0,3,-2])

sum = vector1 + vector2
print(sum)
```

```python
#plotting the vector
plt.quiver(0,0,2,3, scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.quiver(0,0,3,-2, scale_units = 'xy', angles = 'xy',scale = 1, color = 'y')
plt.quiver(0,0,5,1, scale_units = 'xy', angles = 'xy', scale = 1 , color = 'g')
plt.xlim(-8,8)
plt.ylim(-8,8)
```

## Subtraction of 2 vectors

```python
vector1 = np.asarray([0,0,2,3])
vector2 = np.asarray([0,0,3,-2])

subtraction = vector1 - vector2
print(subtraction)
```

```python
#plotting the vector
plt.quiver(0,0,2,3, scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.quiver(0,0,3,-2,scale_units = 'xy', angles = 'xy', scale = 1, color = 'y')
plt.quiver(0,0,-1,5, scale_units = 'xy', angles = 'xy', scale = 1, color = 'g')
plt.xlim(-8,8)
plt.ylim(-8,8)
```

## Multiplying a vector by a scalar

```python
#Multiplying with positive scaler
vector1 = np.asarray([0,0,2,3])
vector2 = 2*vector1
print(vector2)
```

```python
#plotting the vector
plt.quiver(0,0,2,3,scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.quiver(0,0,4,6, scale_units = 'xy', angles = 'xy', scale = 1, color = 'y')
plt.xlim(-8,8)
plt.ylim(-8,8)
```

```python
# multiplying with negative scaler
vector1 = np.asarray([0,0,2,3])
vector2 = -0.5*vector1
print(vector2)
```

```python
#plotting the vector
plt.quiver(0,0,2,3, scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.quiver(0,0,-1,-1.5, scale_units = 'xy', angles = 'xy', scale = 1, color = 'y')
plt.xlim(-4,4)
plt.ylim(-4,4)
```

## Dot Product of two vectors

```python
a = np.array([2,3])
b = np.array([4,4])
a_dot_b = np.dot(a,b)
print(a_dot_b)
```

```python
# three dimensional
c = np.array([40,22,55])
d = np.array([56,33,74])
c_dot_d = np.dot(c,d)
print(c_dot_d)
```

## Cross Product of two vectors

```python
a = np.array([2,3])
b = np.array([4,4])
a_cross_b = np.cross(a,b)
print(a_cross_b)
```

```python
# three dimensional
c = np.array([5,10,25])
d = np.array([23,13,7])
c_cross_d = np.cross(c,d)
print(c_cross_d)
```

## Projection of "a" vector on "v" vector

projection of vector.png

```python
a = np.array([2,5])
v = np.array([8,-6])

#magnitude of "v" vector
magnitude_of_v = np.sqrt(np.sum(v**2))

#projection of "a" on "v"
proj_a_on_v = (np.dot(a,v)/magnitude_of_v**2)*v

print('Projection of "a" vector on "v" vector = ', proj_a_on_v)
```

```python
a = np.array([23, 45, 62])

v = np.array([45, 82, 67])

# magnitude of "v" vector
magnitude_of_v = np.sqrt(np.sum(v**2))

proj_of_a_on_v = (np.dot(a,v)/magnitude_of_v**2)*v

print('Projection of a vector on v vector = ', proj_of_a_on_v)
```
